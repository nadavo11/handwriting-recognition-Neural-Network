```python
import tensorflow as tf
import tensorflow.keras as tk
import numpy as np
import matplotlib.pyplot as plt
import math
import tensorflow as tf
```

```python
mnist = tk.datasets.mnist
(x_train, y_train), (x_test, y_test) = mnist.load_data()

assert x_train.shape == (60000, 28, 28)
assert x_test.shape == (10000, 28, 28)
assert y_train.shape == (60000,)
assert y_test.shape == (10000,)

num_classes = 10

x_train = np.expand_dims(x_train, -1)
x_test = np.expand_dims(x_test, -1)

y_train = tk.utils.to_categorical(y_train, num_classes)
y_test = tk.utils.to_categorical(y_test, num_classes)

input_shape= x_test[1].shape
```

```python
print(x_test.shape)
```

# Initialize the model 
we would need an image proccessing CNN

```python
from tensorflow.keras.layers import Conv2D
from tensorflow.keras.layers import MaxPooling2D
from tensorflow.keras.layers import Flatten
from tensorflow.keras.layers import Dense

# init
model = tk.Sequential()

model.add(tk.Input(shape=input_shape))
# feature extraction
model.add(Conv2D(32,(3,3),input_shape=(128,128,1),activation="relu"))


# maximizing feature map
model.add(MaxPooling2D(pool_size=(2,2)))

def downsample(size):
    i = 128
    while i<=size:
        
        # downsample & maxpooling iteration
        model.add(Conv2D(i,(3,3),activation="relu"))
        model.add(MaxPooling2D(pool_size=(2,2)))

        i*=2

downsample(256)

#flatten to a 1D array
model.add(Flatten())
model.add(Dense(units= 1024, activation='relu',))
model.add(Dense(10, activation='softmax'))
```

# Initialize the model 
we would need an image proccessing CNN

```python
model.compile(optimizer='adam', loss = 'categorical_crossentropy',metrics = ['accuracy'])
model.summary()
```

```python
history = model.fit(x_train,y_train, epochs=15, validation_data=(x_test,y_test))
```
